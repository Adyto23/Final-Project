```python

# Function to extract text from a row
def extract_text(row):
    pattern = r"\((.*?)\)"
    match = re.search(pattern, row)
    if match:
        return match.group(1)
    else:
        return ""
```

```python
df["media_condition"] = df["media_condition"].apply(extract_text)
df["sleeve_condition"] = df["sleeve_condition"].apply(extract_text)

```

```python
df["media_condition"]=df["media_condition"].str.split("or").str[0]
```

```python
df["sleeve_condition"]=df["sleeve_condition"].str.split("or").str[0]
```

```python
df1=df 
df1[["label_name", "label_code"]] = df1["label"].str.extract(r'(.*?)(.*)')
# Remove leading/trailing whitespaces in the label name and code columns
df1["label_name"] = df1["label_name"].str.strip()
df1["label_code"] = df1["label_code"].str.strip()
# Drop the original "label" column
df1 = df1.drop("label", axis=1)
# Drop duplicates based on the "label_name" column
df1 = df1.drop_duplicates(subset="label_name")
```
