```python

# Function to extract text from a brackets in column

def extract_text(row):
    pattern = r"\((.*?)\)"
    match = re.search(pattern, row)
    if match:
        return match.group(1)
    else:
        return ""
```

```python
## extract from brakets in column "media_condition" might end up with blank value will fill out with null or with no cover 
df["media_condition"] = df["media_condition"].apply(extract_text)

```
