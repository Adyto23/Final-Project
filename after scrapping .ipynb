The R2 Score, also known as the coefficient of determination, is a measure of how well the model explains the variation in the data. It ranges from 0 to 1, with a higher score indicating a better fit. An R2 score of 0.48 indicates that the model explains 48% of the variation in the data, while the remaining 52% is due to other factors that the model does not account for.

# ensemble_prediction

```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_squared_error




# Train the models
lr_model = LinearRegression().fit(X_train, y_train)
rf_model = RandomForestRegressor().fit(X_train, y_train)
gb_model = GradientBoostingRegressor().fit(X_train, y_train)

# Combine the models
def ensemble_predict(X):
    return (lr_model.predict(X) + rf_model.predict(X) + gb_model.predict(X)) / 3

# Evaluate the ensemble
y_pred = ensemble_predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print(f"Ensemble R2-Score: {r2:.4f}")
print(f"Ensemble MSE: {mse:.4f}")
print(f"Ensemble RMSE: {rmse:.4f}")


## when we combine the three models it does give us the best result
```

# Plotting in order to find the max_depth that can give us the best results 

```python
import matplotlib.pyplot as plt

# Define a range of max_depth values to test
max_depths = range(1, 30)

# Create empty lists to store the R2 scores for each max_depth
train_scores = []
test_scores = []

# Loop over each max_depth value
for max_depth in max_depths:
    # Define the model with the current max_depth value
    model = RandomForestRegressor(n_estimators=100, max_depth=max_depth)
    
    # Train the model on the training data
    model.fit(X_train, y_train)
    
    # Make predictions on the training and testing data
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # Calculate the R2 scores for the training and testing data
    train_score = r2_score(y_train, y_train_pred)
    test_score = r2_score(y_test, y_test_pred)
    
    # Append the scores to the corresponding lists
    train_scores.append(train_score)
    test_scores.append(test_score)

# Plot the R2 scores as a function of max_depth
plt.plot(max_depths, train_scores, label='Train')
plt.plot(max_depths, test_scores, label='Test')
plt.xlabel('max_depth')
plt.ylabel('R2 score')
plt.legend()
plt.show()


### seem like the good depth will be between 15-30
### as we see the best value will be between 23-25
```
