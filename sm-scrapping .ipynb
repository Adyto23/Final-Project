```python
import re

# Input list of label texts
label_text =country_text
# Regular expression pattern to match common music formats
format_pattern = r"(album|12\"|7\"|compilation|lp|ep|10\")"

# Extract format information using regex
extracted_formats = []
for text in label_text:
    format_match = re.search(format_pattern, text, re.IGNORECASE)
    if format_match:
        extracted_format = format_match.group(0)
        extracted_formats.append(extracted_format)

# Remove duplicates and print the extracted formats
print("Extracted Formats:", extracted_formats)
```

```python
def extract_text(row):
    pattern = r"\((.*?)\)"
    match = re.search(pattern, row)
    if match:
        return match.group(1)
    else:
        return ""
```

```python
df["media_text"] = df["media_text"].apply(extract_text)
df["sleeve_text"] = df["sleeve_text"].apply(extract_text)
```
